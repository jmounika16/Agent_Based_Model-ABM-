## Methodology

```python
import numpy as np           #Include all requried imports  
import pandas as pd           
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from heatmap import heatmap, corrplot
```

```python
#We are considering data from the latest year
#Open 2019.csv file and print it
data_2019=pd.read_csv('2019.csv')
data_2019
```

In the above cell, overall rank is given to countries(ex:Finland is the most happiest country in the year 2019) based on the score which was calculated considering all the factors given.

```python
#Remove column called Generosity
#This is feature that varies a lot. It does help people of certain countries to stay happy but the values can be inaccurate.
data_2019=data_2019.drop(columns=['Generosity'])
```

```python
#summarizing data
data_2019.describe() 
```

```python
data_2019
```

```python
countries_list=list(data_2019['Country or region'].unique()) #creating a list of unique countries in the dataset
```

We can see the data and say which country is most happiest but if we want to compare two countries we need something called the happiness ratio. 

```python
happiness_ratio=[]                                          #https://www.kaggle.com/harumasasuzuki/happiness
for country in countries_list:
    Country=data_2019[data_2019['Country or region']==country] 
    ratio=sum(Country['Score'])/len(countries_list)            #We are calculating ratio by using mean
    happiness_ratio.append(ratio)                              #appending it to the empty list
```

```python
#creating a new data set with two cloumns containing the names of countries and the happiness ratio as score
data=pd.DataFrame({'Country or region':countries_list,'Score':happiness_ratio}) 
index1=data['Score'].sort_values(ascending=False).index.values #sorting the countries based on the scores in descending order
new_data=data.reindex(index1)                                  #re-ordering the data
```

```python
new_data
```

```python
plt.figure(figsize=(25,15))         #setting the size for the bar graph tofit properly                               #https://www.kaggle.com/harumasasuzuki/happiness
sns.barplot(x=new_data['Country or region'][:25], y=new_data['Score'][:25]) #considering only first 25 countries for better visualization.
plt.xlabel('Country or region') #labeling
plt.ylabel('Score')
plt.title('Score Vs Countries')
```

Now let's talk about how the above mentioned factors are related to score.In the below cell, you can see a correlation matrix. It basically tells the correlation between different columns. We can calculate this on a scale of -1 to 1.Though this topic is not taught in class yet,this is one of the easiest and most accurate ways to see which two columns depend on each other and by how much.

We need to do this to check if happiess ratio a.k.a score has actually has something to do with GDP or healthy life expectancy etc.

### Explanation:
The names of all the columns are given on two sides of the square. And also there is a value index strip on the right. Now for example let's take second the row in the square which represents 'score'. First block in that row says -0.99 ,and if you see the title vertically down for that block it said "Overall rank". That means these two values are very less correlated. If one increases the other decreases.

Now go back to the second block in the second row. It said 1. Check the title for that block vertically down, it said score. That means we are comparing two same values so they are perfectly  correlated.

Now see the third block in that second row. It said 0.79. Vertically down it said 'GDP per capita'. This is a positive value and is close to 1,that means these two values score and GDP have good correlation. If the GDP increases the score will also increase and the people will be more happy in that country.

Now see the fourth block, it has the value 0.78. and the title vertically down is 'social support'. It is a positive value and is close to one so the values score and social support have good  correlation. If there is more social support in the country the score a.k.a happiness ratio also increases which leads to a happy country. The same logic applies to next block as well which talks about healthy life expectancy.

Now let's take a look at the sixth block. it has a value of 0.57 and vertically down we have 'Freedom to make life choices'. Although this is still a positive value and there is some correlation between score and Freedom to make life choices, it isn't as much as the previous discussed factors like GDP or healthy life expectancy.

The last block has the value of 0.39.Vertically down we have perceptions of corruptions. Hence the value is close to negative one than plus one, we could say there isn't much correlation between score a.k.a happiness ratio and perceptions of corruption.That means citizen's happiness is not very much dependent on this factor.


```python
correlation_mat = data_2019.corr()
sns.heatmap(correlation_mat, annot = True)
plt.title("Correlation matrix of happiness data")
plt.show()
```

```python
## Conversion of the matrix to 1-D values
#we are unpacking all the values in the above matrix for a better idea.
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
#Below are the correlation co-efficients for variables in the correlation matrix.
```

In the above cell, if you see the values to the right of the score, you can find the same stuff we saw previously in the matrix but if you wanna compare different columns this is a better way to visualise.

### Till now we saw the happiness ratio of different countries and how much a particular factor correlates with the happiness ratio of a country. Now let's visualize them.

This plot is a proof that GDP and score correlate as we told in the previous matrix. Now we can observe the trends of it. And we can also see that as score is increasing GDP per capita is increasing linearly.i.e if score is high, GDP is also high and it's a rich country where people are happy.


```python
fig,ax=plt.subplots()                    #plotting score on x-axis and GDP per capita on y-axis
data_2019.plot(x='Score', y = 'GDP per capita', ax=ax)
plt.title('Score Vs GDP trend')
```

Another linearly increasing plot. People in places where the healthy life expectancy is high , score is also high, in simpler terms where people live longer with good health there they are more happy.

```python
fig,ax=plt.subplots()                                        #plotting score on x-axis and healthy life style on y-axis                 
data_2019.plot(x='Score', y = 'Healthy life expectancy', ax=ax)
plt.title('Score Vs healthy life expectancy trend')
```

Same logic as previous one. In countries where there is strong communal strength and where people support each other the happiness ratio is also high .

```python
fig,ax=plt.subplots()                                  #plotting score on x-axis and social support on y-axis
data_2019.plot(x='Score', y = 'Social support', ax=ax)
plt.title('Score Vs Social support trend')
```

This plot is a little different from the previous ones. In the trends of the graph we can see larger variations because the correlation between freedom to make life choices and score is not so high. That means though the factor freedom to make life choices is very important in calculating happiness ratio, it does not contribute as much as the previous factors.

```python
fig,ax=plt.subplots()     #plotting score on x-axis and Freedom to make life choices on y-axis
data_2019.plot(x='Score', y = 'Freedom to make life choices', ax=ax)
plt.title('Score Vs Freedom to make life choices trend')
```

Here it is very clear that the correlation between perceptions of corruption and score is very less. Though corruption is a very crucial factor to consider it is not contributing much in deciding the happiness ratio of people.

```python
fig,ax=plt.subplots()     #plotting score on x-axis and perceptions of corruption on y-axis
data_2019.plot(x='Score', y = 'Perceptions of corruption', ax=ax)
plt.title('Score Vs Perception of corruption trend')
```

Lastly we have overall rank vs score. Don't get confused by this plot. If you observe carefully you can find that as the score is increasing the overall rank is getting lower. That means as happiness ratio is increasing the overall rank is getting lower(here getting lower means a good thing for example Finland's score is the highest and it's overall rank is the lowest (a.k.a first place of all countries) 

```python
fig,ax=plt.subplots()          #plotting score on x-axis and overall rank on y-axis
data_2019.plot(x='Score', y = 'Overall rank', ax=ax)
plt.title('Score Vs Overall Rank')
```
